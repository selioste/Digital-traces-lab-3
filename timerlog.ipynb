```python
# ouvrir le fichier
with open('text_shakespear.txt', 'r') as f:
  # lire le contenu du fichier
  filename = f.read()
```

```python
def compter_mots_dictionnaire(texte):
    compteur = {}
    mots = texte.split()
    for mot in mots:
        if mot in compteur:
            compteur[mot] += 1
        else:
            compteur[mot] = 1
    return compteur
```

```python
from collections import Counter

def compter_mots_counter(texte):
    mots = texte.split()
    return Counter(mots)
```

```python
import time

def mesurer_temps(fonction, texte):
    debut = time.time()
    resultat = fonction(texte)
    fin = time.time()
    duree = fin - debut
    return duree

# Exemple d'utilisation
temps_dictionnaire = mesurer_temps(compter_mots_dictionnaire, filename)
temps_counter = mesurer_temps(compter_mots_counter, filename)

print(f"La fonction a mis {temps_dictionnaire:.2f} secondes à s'exécuter.")
print(f"La fonction a mis {temps_counter:.2f} secondes à s'exécuter.")
```

```python
import matplotlib.pyplot as plt

def afficher_graphique(temps_dictionnaire, temps_counter):
    # Créez une figure et un graphique
    fig, ax = plt.subplots()

    # Ajoutez les données au graphique
    ax.plot(temps_dictionnaire, label="Dictionnaire")
    ax.plot(temps_counter, label="Counter")

    # Ajoutez une légende et affichez le graphique
    plt.legend()
    plt.show()
```

```python
temps_dictionnaire = []
temps_counter = []

for i in range(100):
    temps_dictionnaire.append(mesurer_temps(compter_mots_dictionnaire, filename))
    temps_counter.append(mesurer_temps(compter_mots_counter, filename))

afficher_graphique(temps_dictionnaire, temps_counter)
```
